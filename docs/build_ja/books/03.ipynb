```正$(8N)角形の場合```は```正$(N)角形の場合```が正しいです

````julia
function sankaku()
    c = sqrt(2)/2
    hankakucos(c) = sqrt((1+c)/2)
    n = 10
    N = 8
    for i=1:n
        c = hankakucos(c)
        N = N*2
        l = sqrt(2 - 2*c)
        println("正$(N)角形の場合：　",N*l/2)
    end
end
sankaku()
````

````julia
function Basel(nc)
    a = 0
    for n=1:nc
        a += 1/n^2
    end
    return sqrt(6a)
end
````

````julia
function Leibniz(nc)
    a = 0
    for n=0:nc
        a += (-1)^n/(2n+1)
    end
    return 4a
end
````

````julia
function test()
    ncs = [10^n for n=0:9]
    bs = []
    ls = []
    for nc in ncs
        b = Basel(nc)
        l = Leibniz(nc)
        push!(bs,abs(π-b)/π)
        push!(ls,abs(π-l)/π)
    end
    println(bs)
    return ncs,bs,ls
end
````

````julia
ncs,bs,ls = test()
````

````julia
plot(ncs,[bs,ls],xscale=:log10,yscale=:log10,markershape = [:circle :star5],label=["Basel" "Leibniz"],xlabel="cutoff num", ylabel="relative error")
````

1刷では```markershape = [:circle,:star5]```となっているが、```markershape = [:circle :star5]```が正しい

````julia
plot(ncs,bs,xscale=:log10,yscale=:log10,markershape = :circle,label="Basel",xlabel="cutoff num",ylabel="relative error")
plot!(ncs,ls,xscale=:log10,yscale=:log10,markershape = :circle,label="Leibniz",xlabel="cutoff num",ylabel="relative error")
````

````julia
function Ramanujan(nc)
    a = 0
    for n=0:nc
        a += factorial(4n)*(1103+26390n) / (4^n*99^n*factorial(n))^4
    end
    return 99^2/(2 * sqrt(2) * a)
end
````

````julia
r = Ramanujan(1)
println(r,"\t",abs(π-r)/π)
````

````julia
r = Ramanujan(4)
println(r,"\t",abs(π-r)/π)
````

````julia
function Kubun(N,x0,x1,f)
    dx = (x1-x0)/N
    a = 0.0
    xn = range(x0,step=dx,length=N)
    for x in xn
        a += f(x)
    end
    return a*dx
end
````

````julia
f(x) = 4/(1+x^2)
N = 10000
p = Kubun(N,0,1,f)
println(p)
````

````julia
N = 1000000
p = Kubun(N,0,1,f)
println(p)
````

````julia
function Daikei(N,x0,x1,f)
    dx = (x1-x0)/N
    a = (f(x0)+f(x1))/2
    xn = range(x0,step=dx,length=N)
    for n=2:N
        x = xn[n]
        a += f(x)
    end 
    return a*dx
end
````

````julia
function MC(N)
    count = 0
    for n=1:N
        x = rand()
        y = rand()
        r = x^2+y^2
        count += ifelse(r > 1,0,1)
    end
    return 4*count/N
end
````

````julia
mutable struct Ball
    v::Float64
    mass::Float64
end
````

````julia
function check_collision!(small::Ball,large::Ball)
    r = large.mass/small.mass
    v = small.v
    V = large.v
    
    if v > V
        V_next = (2v+(r-1)V)/(r+1)
        v_next = ((1-r)v+2r*V)/(r+1)
        collision = true
        
        small.v = v_next
        large.v = V_next
        return collision
    elseif v < 0
        collision = true
        small.v = -v
        return collision
    else
        collision = false
        return collision
    end
end
````

````julia
function ballcollision(N)
    v0 = 0.0
    V0 = -1.0
    m = 1
    M = m*100^N
    smallball = Ball(v0,m)
    largeball = Ball(V0,M)
    collision = true
    count = 0
    while collision
        collision = check_collision!(smallball,largeball)
        if collision 
            count += 1
        end
    end
    return count/10^N
end
````

````julia
N = 5
p = ballcollision(N)
println(p)
````

````julia
mutable struct Ball
    v::Float64
    x::Float64
    v_history::Array{Float64,1}
    x_history::Array{Float64,1}
    mass::Float64
    radius::Float64
    
    function Ball(v,x,m,radius)
        v_history = Float64[]
        x_history = Float64[]
        push!(x_history,x)
        push!(v_history,v)
        return new(v,x,v_history,x_history,m,radius)
    end
end
````

````julia
function update!(ball::Ball,x,v)
    ball.x = x
    ball.v = v
    push!(ball.x_history,x)
    push!(ball.v_history,v)
    return
end
````

````julia
function get_xv(ball::Ball)
    return ball.x,ball.v
end
````

````julia
function check_collision!(small::Ball,large::Ball)
    r = large.mass/small.mass
    x,v = get_xv(small)
    X,V = get_xv(large)
    if v > V
        V_next = (2v+(r-1)V)/(r+1)
        v_next = ((1-r)v+2r*V)/(r+1)
        collision = true
        update!(small,x,v_next)
        update!(large,X,V_next)
        return collision
    elseif v < 0
        collision = true
        update!(small,x,-v)
        update!(large,X,V)
        return collision
    else
        collision = false
        return collision
    end
end
````

````julia
function ball_collision_time(small::Ball,large::Ball)
    t = (small.x+small.radius - (large.x - large.radius))/(large.v - small.v)
end
````

````julia
function wall_collision_time(small::Ball)
    t = -(small.x-small.radius)/small.v
end
````

````julia
function collision_time(small::Ball,large::Ball)
    t_ball = ball_collision_time(small,large)
    t_wall = wall_collision_time(small)
    t = ifelse(t_ball > t_wall && t_wall > 0, t_wall,t_ball)
    return t
end
````

````julia
function check_collision!(small::Ball,large::Ball,timeseries)
    r = large.mass/small.mass
    x,v = get_xv(small)
    X,V = get_xv(large)
    told = timeseries[end]
    if v > V
        t = collision_time(small,large)
        x += t*v
        X += t*V
        V_next = (2v+(r-1)V)/(r+1)
        v_next = ((1-r)v+2r*V)/(r+1)
        collision = true
        update!(small,x,v_next)
        update!(large,X,V_next)
        push!(timeseries,told+t)
        return collision
    elseif v < 0
        t = wall_collision_time(small)
        x += t*v
        X += t*V
        collision = true
        update!(small,x,-v)
        update!(large,X,V)
        push!(timeseries,told+t)
        return collision
    else
        collision = false
        return collision
    end
end
````

````julia
function ballcollision(N)
    v0 = 0.0
    V0 = -1.0
    m = 1
    M = m*100^N
    radius_small = 0.1
    radius_large = 0.1*100^(N/3)
    X0 = 2 + radius_large
    x0 = 1
    smallball = Ball(v0,x0,m,radius_small)
    largeball = Ball(V0,X0,M,radius_large)
    timeseries = Float64[]
    push!(timeseries,0)
    collision = true
    count = 0
    while collision
        collision = check_collision!(smallball,largeball,timeseries)
        if collision 
            count += 1
        end
    end
    make_anime(smallball,largeball,timeseries,N)
    return count/10^N
end
````

````julia
function get_xv(ball::Ball,ith)
    return ball.x_history[ith],ball.v_history[ith]
end
````

````julia
using Plots
function make_anime(small::Ball,large::Ball,timeseries,N)
    anim = Animation()
    ts = range(0.00001,timeseries[end]+0.1,length=200)
    θ = range(0,2π,length=100)
    xcirc = small.radius*cos.(θ)
    ycirc = small.radius*sin.(θ)
    Xcirc = large.radius*cos.(θ)
    Ycirc = large.radius*sin.(θ)
    for t in ts
        ith = searchsortedfirst(timeseries,t)-1
        t0 = timeseries[ith]
        x,v = get_xv(small,ith)
        X,V = get_xv(large,ith)
        dt = t-t0
        x += v*dt
        X += V*dt
        count = ith-1
        plt = plot(x .+ xcirc,ycirc,label="count = $count pi = $(count/10^N)",
            xlims=(0,10),ylims=(-2,2),aspect_ratio = 1)
        plt = plot!(X .+ Xcirc,Ycirc,label=nothing,xlims=(0,10),
        ylims=(-2,2),aspect_ratio=1)
        frame(anim,plt)
    end
    gif(anim,"test_$(N).gif",fps = 15)
end
    
````
